uniform distribution and its visualization

for  cases having equally likely cases

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random
import seaborn as sns
# probability of a coin flip
x= random.uniform(size=8)
sns.distplot(x)
plt.show()
```

```python
# probability of a dice roll
x= random.uniform(size=(4,8))
sns.distplot(x)
plt.show()
```
