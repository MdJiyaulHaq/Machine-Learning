# scatter plot

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('dirtydata.csv')
df.dropna(inplace=True)

df.plot(kind = 'scatter', x = "Duration", y= 'Calories')
plt.show() # duration and calories have good correlation 
```
