# fixing the wrong input data 

```python
import pandas as pd

df = pd.read_csv('dirtydata.csv')
df.loc[7,'Duration'] = 43 # this will replace the wrong data ie. 450 at ind3x 7 with 43
df
```

for larger data sets you can create some rules, e.g. set some boundaries for legal values, and replace any values that are outside of the boundaries

```python
import pandas as pd

df = pd.read_csv('dirtydata.csv')

for x in df.index:
  if df.loc[x, "Duration"] < 33 or df.loc[x, "Duration"] >66:
    df.loc[x, "Duration"] = 45

df
```

if the data set is to large and you don't really know what to replace inplace of the wrong data, you can simply delete that row

```python
import pandas as pd

dfc= pd.read_csv('dirtydata.csv')
df=dfc
for x in df.index:
    if df.loc[x,'Duration'] > 66: 
        df.drop(x)
        
df
```
